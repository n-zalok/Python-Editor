```python
import pandas as pd

df = pd.read_csv("../data/exploration.csv")
```

Splitting the data randomly may lead to data leakage since some developers will appear in both training and test sets. Some developers may be more skilled at writing high quality code than others. If a developer appears in both the training and the test set, a model could successfully predict the performance of his/her code simply by successfully identifying the developer.

```python
from sklearn.model_selection import train_test_split

random_train, random_test = train_test_split(df, test_size=0.3, random_state=0)
train_developers =  set(random_train.apply(lambda row: row["repo_name"].split("/")[0], axis=1))
test_developers  =  set(random_test.apply(lambda row: row["repo_name"].split("/")[0], axis=1))

print(f"{len(train_developers.intersection(test_developers))} developers appear in both sets")
```

Splitting by the developer solves this problem

```python
import sys
sys.path.append("..")

from python_editor.data_processing import split_by_developer

train, test = split_by_developer(df, test_size=0.3, random_state=0)
train_developers =  set(train["developer"])
test_developers  =  set(test["developer"])

print(f"{len(train_developers.intersection(test_developers))} developers appear in both sets")
```
